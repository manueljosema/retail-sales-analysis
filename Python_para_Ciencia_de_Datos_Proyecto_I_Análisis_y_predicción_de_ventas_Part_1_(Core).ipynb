```python
# Paso 1: Carga de Datos
# Cargar los datos desde el archivo CSV usando pandas
file_path = '/content/retail_sales_dataset.csv'
try:
    df = pd.read_csv(file_path)
    print("Datos cargados exitosamente.")
except Exception as e:
    print(f"Error al cargar los datos: {e}")

```

```python
# Paso 2: Preprocesamiento Básico
# Eliminar filas que contengan valores NaN y asegurarse de que los datos estén limpios
df_cleaned = df.dropna()
print("Datos preprocesados.")
```

### 2) Exploración de Datos:
* Calcula el total de ventas por categoría de producto.
* Calcula el promedio de ventas diarias por categoría de producto.
* Identifica las categorías de productos con mayores y menores ventas.

```python
# Total de ventas por categoría de producto
total_ventas_por_categoria = df_cleaned.groupby('Product Category')['Price per Unit'].sum()
print("Total de ventas por categoría:\n", total_ventas_por_categoria)

# Promedio de ventas diarias por categoría de producto
promedio_ventas_diarias = df_cleaned.groupby(['Date','Product Category','Quantity'])['Price per Unit'].mean()
print("\nPromedio de ventas diarias por categoría:\n", promedio_ventas_diarias)
```

```python
# Categorías con mayores y menores ventas
ventas = {
    "Beauty": 56505,
    "Clothing": 61175,
    "Electronics": 62210
}
def mayor_y_menor_venta(ventas):

  mayor_venta = max(ventas, key=ventas.get)
  menor_venta = min(ventas, key=ventas.get)

  return mayor_venta, ventas[mayor_venta], menor_venta, ventas[menor_venta]

resultado = mayor_y_menor_venta(ventas)
print("El producto más vendido fue:", resultado[0], "con", resultado[1], "unidades vendidas.")
print("El producto menos vendido fue:", resultado[2], "con", resultado[3], "unidades vendidas.")
```

### 3) Manipulación de Datos:
* Filtra los datos para mostrar solo las ventas de una categoría de producto específica.
* Realiza operaciones de suma, resta, multiplicación y división en los datos para obtener estadísticas adicionales.

```python
# Crear df_cleaned eliminando filas con valores NaN
df_cleaned = df.dropna()
# Filtrar datos de una categoría específica
categoria_especifica = 'Beauty'
ventas_categoria_especifica = df_cleaned[df_cleaned['Product Category'] == categoria_especifica]

# Operaciones matemáticas básicas para estadísticas adicionales
suma_ventas = df_cleaned['Total Amount'].sum()
resta_ventas = df_cleaned['Total Amount'].diff()  # Diferencias entre ventas sucesivas
multiplicacion_ventas = df_cleaned['Total Amount'].prod()
division_promedio = df_cleaned['Total Amount'].mean()

print("Ventas de la categoría 'Beauty':\n", ventas_categoria_especifica)
print("\nSuma de las ventas:", suma_ventas)
print("Diferencias entre ventas sucesivas:\n"), resta_ventas
print("\nMultiplicación de las ventas:", multiplicacion_ventas)
print("\nPromedio de las ventas:", division_promedio)
```
